Import Pandas 

```python
import pandas as pd
```

Reading File and checking Date header

```python
data = pd.read_csv("NIFTY50-20221020121500000_20220902122500000.csv",parse_dates=True)
data["Date"]
```

rsplit Date column to see which items to save.

```python
data["Date"].str.rsplit(expand=True)
```

Saving Day, Month, Date(Dt), Year and Timestamps for each 5 min candle.

```python
data["Day"] = data["Date"].str.rsplit(expand=True)[0]
data["Month"] = data["Date"].str.rsplit(expand=True)[1]
data["Dt"] = data["Date"].str.rsplit(expand=True)[2]
data["Year"] = data["Date"].str.rsplit(expand=True)[3]
data["Time"] = data["Date"].str.rsplit(expand=True)[4]
data
```

Saving OpenClose and HighLow diffs.

```python
data["OpenClose"] = ((1 - data["Open"]/data["Close"])*100).round(4)
data["HighLow"] = ((1 - data["High"]/data["Low"])*100).round(4)
```

Taking one of the Thursday for analysis

```python
data.loc[data["Dt"] == "08"]
```

printing cumsum of one of the Thursday

```python
data["OpenClose"].loc[data["Dt"] == "08"].cumsum()
```

Save Newer File with additional headers 

```python
data.to_csv("nifty50-5minCandle-details.csv",index=False)
```
